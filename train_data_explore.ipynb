# Create a map with all the stations

```python
import pandas as pd
import folium

# Load your data
stations = pd.read_csv('data/train_stations.csv', delimiter=";", header=0)
train_data_small = pd.read_csv("data/train_data_small.csv", delimiter=",", header=0)

merged_df = stations.merge(train_data_small, on="OPERATING_POINT_ABBREVIATION")

merged_df = merged_df[["OPERATING_POINT_ABBREVIATION","BP_Name", "Latitude","Longitude"]].drop_duplicates(subset=["OPERATING_POINT_ABBREVIATION"])

import plotly.express as px

# Assuming 'stations' DataFrame has columns: 'latitude', 'longitude', and 'station_id'
fig = px.scatter_mapbox(
    merged_df,
    lat='Latitude',
    lon='Longitude',
    hover_name='BP_Name',  # or another identifier column
    zoom=10,
    height=600
)
fig.update_layout(
    mapbox_style="open-street-map",
    margin={"r":0,"t":0,"l":0,"b":0}
)
fig.show()
```
